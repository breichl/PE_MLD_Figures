```python
import matplotlib.pyplot as plt
from oceanmixedlayers import oceanmixedlayers
from Paths import Figure_Save_Path
plt.rcParams["font.family"] = "Serif"


#Specify the 3 idealized profiles
A=oceanmixedlayers.column(kind='idealized',idealized_type='linear',EOS='Linear',
                          T0=20,dTdz=(0.03/100)/0.2,S0=35,nz=1000,Dpt=100)
B=oceanmixedlayers.column(kind='idealized',idealized_type='two-layer',EOS='Linear',
                          T0=20,Tb0=20-(0.03)/0.2,nz=1000,Dpt=100)
C=oceanmixedlayers.column(kind='idealized',idealized_type='two-layer',EOS='Linear',
                          T0=20,Tb0=20,dTdz=(0.03/(100*0.1))/0.2,nz=1000,Dpt=100,mixedfrac=0.9)


#Create the figure
F,AX = plt.subplots(1,1,figsize=(4,5))
AX.plot(A.prho-A.prho[0],A.zc,'r--',linewidth=3,label='$\phi$={:3.2f} '.format(A.pe_anomaly(Dpt=-100)[0])+'J m$^{-2}$')
AX.plot(B.prho-B.prho[0],B.zc,'b-.',linewidth=3,label='$\phi$={:3.2f} '.format(B.pe_anomaly(Dpt=-100)[0])+'J m$^{-2}$')
AX.plot(C.prho-C.prho[0],C.zc,'g:',linewidth=3,label='$\phi$={:3.2f} '.format(C.pe_anomaly(Dpt=-100)[0])+'J m$^{-2}$')
AX.set_xlabel(r'$\Delta\rho_\theta$ [kg/m$^3$]')
AX.set_ylabel('$z$ [m]')
AX.set_title(r'$\Delta\rho_\theta$ profile'+'\n PE Anomaly ($\phi$)')
AX.legend(fontsize=12)
AX.set_ylim(-102,0)
AX.set_xlim(-.003,.031)

F.tight_layout()

F.savefig(Figure_Save_Path+'/Figure1_Idealized_profiles.png');
```
